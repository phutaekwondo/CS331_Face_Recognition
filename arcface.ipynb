```python
from arcface import ArcFaceModel
import cv2
import numpy as np
import os
```

```python
model = ArcFaceModel()
model.summary()
```

```python
list_image = []

image1 = cv2.imread("img/cuong.jpg")
image2 = cv2.imread("img/cuong2.jpg")

list_image.append(image1)
list_image.append(image2)
```

```python
embedding1 = model.predict(image1)
embedding2 = model.predict(image2)

print(model.cosine(embedding1, embedding2))
```
