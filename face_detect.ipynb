# face detect (mtcnn)

```python
from mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf

# tf.debugging.set_log_device_placement(True)

img_path = 'img/face.jpg'
img = cv2.imread(img_path)
detector = MTCNN()
result = detector.detect_faces(img)
print(result)
```

```python
#draw the bounding box
bounding_box = result[0]['box']
keypoints = result[0]['keypoints']
img_bb = img.copy()
BLUE = (255,0,0)
cv2.rectangle(img_bb,
                (bounding_box[0], bounding_box[1]),
                (bounding_box[0]+bounding_box[2], bounding_box[1] + bounding_box[3]),
                BLUE,
                2)
cv2.circle(img_bb,(keypoints['left_eye']), 2,BLUE, 2)
cv2.circle(img_bb,(keypoints['right_eye']), 2,BLUE, 2)
cv2.circle(img_bb,(keypoints['nose']), 2,BLUE, 2)
cv2.circle(img_bb,(keypoints['mouth_left']), 2,BLUE, 2)
cv2.circle(img_bb,(keypoints['mouth_right']), 2,BLUE, 2)

plt.imshow(cv2.cvtColor(img_bb, cv2.COLOR_BGR2RGB))
plt.show()
```
